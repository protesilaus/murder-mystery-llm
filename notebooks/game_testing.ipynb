```python
# Core imports
from mmllm.game.engine import GameEngine
from mmllm.game.loop import GameLoop
from mmllm.core.types import Phase, ActionType, ActionResponse, KillAction, SpeakAction, PassAction, AgentAction
from mmllm.core.game_config import GameTypeConfig
from mmllm.agents.scripted_agent import ScriptedAgent
from mmllm.game.rules import legal_actions, winner
import json

print("Imports successful!")
```

## 1. Create a Game

Let's create a basic game with the default configuration.

```python
# Create a basic game
engine = GameEngine(
    game_id="test_game_001",
    player_ids=["alice", "bob", "charlie", "diana", "eve"],
    murderer_id="alice",
)

# Start the game
engine.start()

print(f"Game created: {engine.runtime.public_state.game_id}")
print(f"Current phase: {engine.current_phase()}")
print(f"Players: {[p.player_id for p in engine.runtime.public_state.players]}")
print(f"Murderer: alice")
```

```python
# ═══════════════════════════════════════════════════════════
# GAME STATUS
# ═══════════════════════════════════════════════════════════

print("┌" + "─" * 58 + "┐")
print(f"│  Game Name:      {engine.runtime.public_state.game_id:<40}│")
print(f"│  Current Phase:  {engine.current_phase().value:<40}│")
print(f"│  Round:          {engine.runtime.public_state.round_num:<40}│")
print("└" + "─" * 58 + "┘")
```

## 2. Run One Round

Execute one complete round (all players act once in the current phase).

```python
# Create agents for each player
agents = {p.player_id: ScriptedAgent(seed=idx) for idx, p in enumerate(engine.runtime.public_state.players)}
# Create game loop
loop = GameLoop(engine, agents)
```

```python
print("Before running round:")
print(f"  Phase: {engine.current_phase().value}")
print(f"  Round: {engine.runtime.public_state.round_num}")
print(f"  Events: {len(engine.runtime.event_history)}")

# Run one complete round
loop.run_round()

print("\nAfter running round:")
print(f"  Phase: {engine.current_phase().value}")
print(f"  Round: {engine.runtime.public_state.round_num}")
print(f"  Events: {len(engine.runtime.event_history)}")
print(f"  Alive: {sum(1 for p in engine.runtime.public_state.players if p.alive)}/{len(engine.runtime.public_state.players)}")
```
